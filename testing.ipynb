```python
df = pd.DataFrame({
    'DateTime': pd.date_range(start='2012-01-01 00:00:00', end='2013-12-31 23:00:00', freq='1h')
})
train_months=21
test_months=3
overlap_months=0
train_test_blocks=blocked_cross_validation_idx(df,train_months,test_months,overlap_months)[0]
```

```python
def get_train_batch_blocks(start,end,batch_size=64):
    blocks=[]
    curr_start=start
    while curr_start<end:
        curr_end=min(curr_start+batch_size-1,end)
        blocks.append((curr_start,curr_end))
        curr_start+=batch_size
    return blocks
```

```python
batch_idxs=get_train_batch_blocks(train_test_blocks[0],train_test_blocks[1])
```
